## Question 1 
Imported the airport data from both csv's using pandas. Then filtered this data and put it into SQL.

```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
airports = pd.read_csv("hw_5_data/top_airports.csv")
locations = pd.read_csv("hw_5_data/ICAO_airports.csv")
```

```python
airports.head()
```

```python
locations.head()
```

```python
airport_loc = []
for airport in airports['IATA']:
    for idx, iata in enumerate(locations['iata_code']):
        if iata == airport:
            airport_loc.append([locations['name'][idx],locations['latitude_deg'][idx],locations['longitude_deg'][idx],locations['iata_code'][idx]])
```

```python
airport_loc
```

```python
latitudes = [item[1] for item in airport_loc]
longitudes = [item[2] for item in airport_loc]
```

```python
airports['Latitude'] = latitudes
airports['Longitude'] = longitudes
airports.drop(columns=['FAA', 'ICAO','Enplanements','Role'])
```

```python
!rm airports.db
connection1 = sqlite3.connect('airports.db')
cursor1 = connection1.cursor()
sql_cmd1 = """CREATE TABLE airports (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    city TEXT,
    airport TEXT,
    IATA TEXT, 
    latitude FLOAT,
    longitude FLOAT)"""
cursor1.execute(sql_cmd1)
connection1.commit()
```

```python
for i, row in airports.iterrows():
    params = (row["City"],row["Airport"],row["IATA"],row["Latitude"],
              row["Longitude"])

    cursor1.execute("INSERT INTO airports" + \
                   "(City,Airport,IATA,Latitude,Longitude)  " + \
                   "VALUES (?, ?, ?, ?, ?)", params)
```

```python
sql_cmd = "SELECT count(*) FROM airports"
cursor1.execute(sql_cmd)
```
